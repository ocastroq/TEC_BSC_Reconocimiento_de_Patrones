## 1. kNN 

##### Se importan las librerias a utilizar: 

```python
from sklearn.datasets import load_wine 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
%matplotlib inline
```

#### Se carga el dataset load_wine

#### Se realiza el split de los datos de training y de test, asignando un 20% de los datos para test y se varian los valores de ramdom_state, de manera que  para buscar la configuracion mas optima:

```python
wine = load_wine()
X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.2, random_state=10)
training_accuracy = []
test_accuracy = []
neighbors_settings = range(1, 10)
for n_neighbors in neighbors_settings:
    # se construye el modelo de clasificacion
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    # se almacena el "training set accuracy"
    training_accuracy.append(clf.score(X_train, y_train))
    # se almacena la "generalization accuracy"
    test_accuracy.append(clf.score(X_test, y_test))
plt.plot(neighbors_settings, training_accuracy, label="training accuracy")
plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
plt.ylabel("Accuracy")
plt.xlabel("N_neighbors")
plt.legend()
```

#### Para este caso se observa que el maximo accuracy para el data set de test es de un 75% aproximadamente con N-neighbors = 4, mientras que el accuracy del dataset de training a este punto es de aproximadamente = 84%

## 3. SVM

### Primeramente, se importa la libreria de SVC y en seguida se se ejecuta la funcion SVC con el fin de obtener el score del Training y el score del Testing:

```python
from sklearn.svm import SVC

X_train, X_test, y_train, y_test = train_test_split(
wine.data, wine.target, random_state=0)
svc = SVC(gamma='auto')
svc.fit(X_train, y_train)
print("Accuracy on training set: {:.2f}".format(svc.score(X_train, y_train)))
print("Accuracy on test set: {:.2f}".format(svc.score(X_test, y_test)))
```

#### Se realiza una grafica que muesta los valores maximos y minimos para el dataset:

```python
plt.plot(X_train.min(axis=0), '*', label="min")
plt.plot(X_train.max(axis=0), '+', label="max")
plt.legend(loc=4)
plt.xlabel("Feature index")
plt.ylabel("Feature magnitude")
plt.yscale("log")
```

#### A continuacion se ejecuta un cambio de escala de manera que sean valores entre 0 y 1:

```python
# compute the minimum value per feature on the training set
min_on_training = X_train.min(axis=0)

# compute the range of each feature (max - min) on the training set
range_on_training = (X_train - min_on_training).max(axis=0)

# subtract the min, and divide by range
# afterward, min=0 and max=1 for each feature
X_train_scaled = (X_train - min_on_training) / range_on_training

print("Minimum for each feature\n{}".format(X_train_scaled.min(axis=0)))
print("Maximum for each feature\n {}".format(X_train_scaled.max(axis=0)))
```

```python
# use THE SAME transformation on the test set,
# using min and range of the training set

X_test_scaled = (X_test - min_on_training) / range_on_training
```

```python
plt.figure()
plt.plot(X_train_scaled.min(axis=0), '*', label="min")
plt.plot(X_train_scaled.max(axis=0), '+', label="max")
plt.legend(loc=5)
plt.xlabel("Feature index")
plt.ylabel("Feature magnitude")
```

#### Para este caso se establece el valor del hiperparametro C de forma automatica: 

```python
svc = SVC(gamma='auto')
svc.fit(X_train_scaled, y_train)
print("Accuracy on training set: {:.3f}".format(
svc.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

#### Posteriormente se asigna un valor de entrada a los hiperparametros: 
##### C: regulacion 
##### gamma: control del ancho del kernel gaussiano 

```python
svc = SVC(C=50, gamma=0.1)
svc.fit(X_train_scaled, y_train)
print("Accuracy on training set: {:.3f}".format(
svc.score(X_train_scaled, y_train)))
print("Accuracy on test set: {:.3f}".format(svc.score(X_test_scaled, y_test)))
```

# Metodos de Ensable

#### Los metodos de de ensamble buscan mejorar el rendimiento de los algoritmos de aprendizaje supervisado a la hora de resolver algun problema 

#### A continuacion se estaran ejecutando metodos de ensamble, por lo que se procede a importar la libreria para este caso ademas de otras librerias relevantes al ejercicio:

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, AdaBoostClassifier, VotingClassifier
```

#### Para esta caso se estara utilizando el set de datos del cual se puede encontrar mayor informacion en el siguiente enlace:

https://archive.ics.uci.edu/ml/datasets/Wine+Quality

```python
wine_quality=pd.read_csv('https://raw.githubusercontent.com/ocastroq/TEC_BSC_Reconocimiento_de_Patrones/main/winequality-red%20(2).csv', sep=';')
```

```python
wine_quality.head()
```

#### Se eligen las variables "x" y "y":

```python
x=np.asanyarray(wine_quality.iloc[:,:-1])
y=np.asanyarray(wine_quality.iloc[:,-1])
```

#### Se realiza el split de los datos de training y test: 

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state=0)
```

#### Se crea un Arbol de decision a raiz del cual se estarian posteriormente ejectuando los modelos de ensamble:

```python
dt=DecisionTreeClassifier()
dt.fit(xtrain,ytrain)
print('Train:', dt.score(xtrain,ytrain))
print('Test:', dt.score(xtest,ytest))
```

#### Metodo de ensamble 1. RandomForest

```python
rf=RandomForestClassifier(n_estimators=65)
rf.fit(xtrain,ytrain)
print('Train:', rf.score(xtrain,ytrain))
print('Test:', rf.score(xtest,ytest))
```

#### Metodo de ensamble 2. Bagging

```python
bg = BaggingClassifier(DecisionTreeClassifier(),max_samples=0.5, max_features=1.0,n_estimators=100)
bg.fit(xtrain,ytrain)
print('Train:', bg.score(xtrain,ytrain))
print('Test:', bg.score(xtest,ytest))
```

#### Metodo de ensamble 3. AdaBoost

```python
adb=AdaBoostClassifier(DecisionTreeClassifier(),n_estimators=10,learning_rate=5)
adb.fit(xtrain,ytrain)
print('Train:', adb.score(xtrain,ytrain))
print('Test:', adb.score(xtest,ytest))
```

#### Metodo de ensamble 4. Voting

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
lr=LogisticRegression(solver='lbfgs', max_iter=15000)
dt=DecisionTreeClassifier()
svm=SVC(kernel='rbf', gamma='scale')
```

```python
evc = VotingClassifier( estimators= [('lr',lr),('dt',dt),('svm',svm)], voting='hard')
evc.fit(xtrain, ytrain)
print('Train:', bg.score(xtrain,ytrain))
print('Test:', bg.score(xtest,ytest))
```
